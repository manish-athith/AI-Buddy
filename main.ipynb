#  Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import random

print("✓ Libraries imported successfully!")
```

#  Create Science Questions

```python
print("CREATING SCIENCE QUESTIONS")
print("=" * 40)

science_questions = []
science_labels = []

science_templates = [
    "Why is the sky {}?",
    "How does {} work?",
    "What is {}?",
    "Tell me about {}",
    "Why do {} happen?",
    "How do {} function?"
]

science_fillers = ["blue", "photosynthesis", "gravity", "electricity", "magnetism", 
                  "rainbows", "seasons", "tides", "volcanoes", "earthquakes"]

# Generate 40 science questions
for i in range(40):
    template = random.choice(science_templates)
    filler = random.choice(science_fillers)
    question = template.format(filler)
    science_questions.append(question)
    
    science_labels.append({
        'age_group': random.choice(['elementary', 'middle_school', 'high_school']),
        'subject': 'science',
        'language': 'english',
        'intent': 'factual_doubt' if random.random() > 0.3 else 'concept_explanation'
    })

print(f"✓ Created {len(science_questions)} science questions")
print("Sample science questions:")
for i in range(3):
    print(f"  {i+1}. {science_questions[i]}")
    print(f"     Labels: {science_labels[i]}")
```

# Create Math Questions

```python
print("\nCREATING MATH QUESTIONS")
print("=" * 40)

math_questions = []
math_labels = []

math_templates = [
    "How do I solve {}?",
    "Explain how to calculate {}",
    "What is the formula for {}?",
    "Help me with this math problem: {}",
    "How to find {}?"
]

math_fillers = ["2x + 5 = 15", "the area of a circle", "quadratic equations", 
               "Pythagorean theorem", "algebraic expressions", "fractions"]

# Generate 35 math questions
for i in range(35):
    template = random.choice(math_templates)
    filler = random.choice(math_fillers)
    question = template.format(filler)
    math_questions.append(question)
    
    math_labels.append({
        'age_group': random.choice(['middle_school', 'high_school']),
        'subject': 'math',
        'language': 'english',
        'intent': 'homework_help'
    })

print(f"✓ Created {len(math_questions)} math questions")
print("Sample math questions:")
for i in range(3):
    print(f"  {i+1}. {math_questions[i]}")
    print(f"     Labels: {math_labels[i]}")
```

```python
print("\nCREATING HISTORY QUESTIONS")
print("=" * 40)

history_questions = []
history_labels = []

history_templates = [
    "What caused {}?",
    "Explain the significance of {}",
    "Why was {} important?",
    "What led to {}?",
    "Describe the events of {}"
]

history_fillers = ["World War II", "the French Revolution", "ancient civilizations", 
                  "the Industrial Revolution", "the Cold War"]

# Generate 30 history questions
for i in range(30):
    template = random.choice(history_templates)
    filler = random.choice(history_fillers)
    question = template.format(filler)
    history_questions.append(question)
    
    history_labels.append({
        'age_group': random.choice(['middle_school', 'high_school']),
        'subject': 'history',
        'language': 'english',
        'intent': 'concept_explanation'
    })

print(f"✓ Created {len(history_questions)} history questions")
print("Sample history questions:")
for i in range(3):
    print(f"  {i+1}. {history_questions[i]}")
    print(f"     Labels: {history_labels[i]}")
```

# Create Language Questions

```python
print("\nCREATING LANGUAGE QUESTIONS")
print("=" * 40)

language_questions = []
language_labels = []

# Let's create separate templates for different types to avoid formatting issues
translation_templates = [
    "How do you say {} in {}?",
    "Translate {} to {}"
]

grammar_templates = [
    "Explain the {} grammar rule",
    "What is the {} rule in grammar?",
    "Help me understand {} grammar"
]

vocabulary_templates = [
    "What does {} mean?",
    "Can you explain what {} means?",
    "Tell me the meaning of {}"
]

# Generate 20 language questions
for i in range(20):
    question_type = random.choice(['translation', 'grammar', 'vocabulary'])
    
    if question_type == 'translation':
        template = random.choice(translation_templates)
        word = random.choice(["hello", "thank you", "goodbye", "water", "food"])
        language = random.choice(["Hindi", "Spanish", "French"])
        question = template.format(word, language)
        
    elif question_type == 'grammar':
        template = random.choice(grammar_templates)
        grammar_point = random.choice(["past tense", "present perfect", "articles", "prepositions", "verb conjugation"])
        question = template.format(grammar_point)
        
    else:  # vocabulary
        template = random.choice(vocabulary_templates)
        word = random.choice(["benevolent", "ephemeral", "resilient", "ambiguous", "paradox"])
        question = template.format(word)
    
    language_questions.append(question)
    
    language_labels.append({
        'age_group': random.choice(['elementary', 'middle_school', 'high_school']),
        'subject': 'language',
        'language': 'english',
        'intent': 'language_help'
    })

print(f"✓ Created {len(language_questions)} language questions")
print("Sample language questions:")
for i in range(3):
    print(f"  {i+1}. {language_questions[i]}")
    print(f"     Labels: {language_labels[i]}")
```

#  Data Preprocessing

```python
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

print("\n" + "=" * 50)

# Let's create features for each category we want to predict
categories = ['age_group', 'subject', 'language', 'intent']

print("Unique values in each category:")
for category in categories:
    unique_vals = df[category].unique()
    print(f"{category}: {list(unique_vals)}")

print("\n" + "=" * 50)

# Prepare features (text) and targets (labels)
X = df['question']  # Our input features (the questions)
y_age = df['age_group']  # Target for age group prediction
y_subject = df['subject']  # Target for subject prediction  
y_language = df['language']  # Target for language prediction
y_intent = df['intent']  # Target for intent prediction

print("✓ Features and targets prepared")
print(f"Features shape: {X.shape}")
print(f"Sample feature: '{X.iloc[0]}'")
print(f"Sample targets:")
print(f"  - Age: {y_age.iloc[0]}")
print(f"  - Subject: {y_subject.iloc[0]}")
print(f"  - Language: {y_language.iloc[0]}")
print(f"  - Intent: {y_intent.iloc[0]}")
```

# Text Vectorization

```python
# We'll use TF-IDF Vectorizer which is better than simple CountVectorizer
vectorizer = TfidfVectorizer(
    max_features=500,  # Consider only top 500 words to keep it simple
    stop_words='english',  # Remove common English words
    ngram_range=(1, 1)  # Consider single words only for now
)

# Fit and transform the text data
print("Fitting vectorizer to our text data...")
X_vectorized = vectorizer.fit_transform(X)

print(f"✓ Text vectorization completed")
print(f"Original text shape: {X.shape}")
print(f"Vectorized shape: {X_vectorized.shape}")
print(f"Number of features (vocabulary size): {len(vectorizer.get_feature_names_out())}")

print("\nSample feature names (first 20):")
print(vectorizer.get_feature_names_out()[:20])
```

# Train-Test Split

```python
# Split data for each target variable (we'll use the same split for all)
test_size = 0.2  # 20% for testing, 80% for training

# Split the data
X_train, X_test, y_age_train, y_age_test = train_test_split(
    X_vectorized, y_age, test_size=test_size, random_state=42
)

# For other targets, we need to use the same indices to maintain consistency
X_train, X_test, y_subject_train, y_subject_test = train_test_split(
    X_vectorized, y_subject, test_size=test_size, random_state=42
)

X_train, X_test, y_language_train, y_language_test = train_test_split(
    X_vectorized, y_language, test_size=test_size, random_state=42
)

X_train, X_test, y_intent_train, y_intent_test = train_test_split(
    X_vectorized, y_intent, test_size=test_size, random_state=42
)

print("✓ Data split completed successfully")
print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
print(f"Feature dimension: {X_train.shape[1]}")

print("\nTraining set distribution (Intent):")
print(pd.Series(y_intent_train).value_counts())

print("\nTesting set distribution (Intent):")
print(pd.Series(y_intent_test).value_counts())
```

# Model Training

```python

# We'll train separate classifiers for each category
classifiers = {}

print("1. Training Age Group Classifier...")
classifiers['age_group'] = RandomForestClassifier(n_estimators=50, random_state=42)
classifiers['age_group'].fit(X_train, y_age_train)
print("   ✓ Age Group Classifier trained")

print("\n2. Training Subject Classifier...")  
classifiers['subject'] = RandomForestClassifier(n_estimators=50, random_state=42)
classifiers['subject'].fit(X_train, y_subject_train)
print("   ✓ Subject Classifier trained")

print("\n3. Training Language Classifier...")
classifiers['language'] = MultinomialNB()
classifiers['language'].fit(X_train, y_language_train)
print("   ✓ Language Classifier trained")

print("\n4. Training Intent Classifier...")
classifiers['intent'] = RandomForestClassifier(n_estimators=50, random_state=42)
classifiers['intent'].fit(X_train, y_intent_train)
print("   ✓ Intent Classifier trained")

print("\n" + "=" * 50)
print(" ALL MODELS TRAINED SUCCESSFULLY!")
```

 # Model Evaluation

```python
# We'll evaluate each classifier and see how well they perform
classifier_names = ['age_group', 'subject', 'language', 'intent']
predictions = {}

print("EVALUATING EACH CLASSIFIER:")
print("=" * 40)

for name in classifier_names:
    print(f"\n{name.upper()} CLASSIFIER:")
    print("-" * 30)
    
    # Get the correct test data for this classifier
    if name == 'age_group':
        y_true = y_age_test
    elif name == 'subject':
        y_true = y_subject_test
    elif name == 'language':
        y_true = y_language_test
    else:  # intent
        y_true = y_intent_test
    
    # Make predictions
    y_pred = classifiers[name].predict(X_test)
    predictions[name] = y_pred
    
    # Calculate accuracy
    accuracy = accuracy_score(y_true, y_pred)
    
    print(f"Accuracy: {accuracy:.2f} ({accuracy * 100:.1f}%)")
    print(f"Correct: {sum(y_true == y_pred)}/{len(y_true)}")
    
    # Show some examples of predictions vs actual
    print("\nSample predictions:")
    for i in range(min(3, len(y_true))):
        print(f"  {i+1}. Predicted: {y_pred[i]:<15} | Actual: {y_true.iloc[i]:<15} | {'✓' if y_pred[i] == y_true.iloc[i] else '✗'}")

print("\n" + "=" * 50)
print(" OVERALL PERFORMANCE SUMMARY")
print("=" * 50)

# Calculate overall accuracy across all classifiers
total_correct = 0
total_predictions = 0

for name in classifier_names:
    if name == 'age_group':
        y_true = y_age_test
    elif name == 'subject':
        y_true = y_subject_test
    elif name == 'language':
        y_true = y_language_test
    else:
        y_true = y_intent_test
    
    y_pred = predictions[name]
    correct = sum(y_true == y_pred)
    total_correct += correct
    total_predictions += len(y_true)
    
    accuracy = accuracy_score(y_true, y_pred)
    print(f"{name.upper():<12}: {accuracy:.2f} ({correct:>2}/{len(y_true):>2} correct)")

overall_accuracy = total_correct / total_predictions
print(f"\n OVERALL ACCURACY: {overall_accuracy:.2f} ({total_correct}/{total_predictions} predictions correct)")
```

# Create Prompt Templates

```python
print("BUILDING PROMPT TEMPLATE SYSTEM")
print("=" * 50)

# Define prompt templates for different combinations of labels
prompt_templates = {
    # Science + Factual Doubt combinations
    ('science', 'factual_doubt', 'elementary'): 
        "Explain this in simple, fun terms for a young child: {question}",
    
    ('science', 'factual_doubt', 'middle_school'): 
        "Provide a clear, factual explanation for a middle school student: {question}",
    
    ('science', 'factual_doubt', 'high_school'): 
        "Give a detailed, scientific explanation with key concepts: {question}",
    
    # Science + Concept Explanation
    ('science', 'concept_explanation', 'elementary'): 
        "Explain this concept like you're talking to a curious 8-year-old. Use simple analogies and examples: {question}",
    
    ('science', 'concept_explanation', 'middle_school'): 
        "Explain this concept clearly with real-world examples suitable for teenagers: {question}",
    
    ('science', 'concept_explanation', 'high_school'): 
        "Provide a comprehensive explanation with scientific principles and applications: {question}",
    
    # Math + Homework Help
    ('math', 'homework_help', 'middle_school'): 
        "Provide step-by-step guidance to solve this math problem. Explain each step clearly: {question}",
    
    ('math', 'homework_help', 'high_school'): 
        "Solve this math problem with detailed working and explain the underlying concepts: {question}",
    
    # History + Concept Explanation
    ('history', 'concept_explanation', 'middle_school'): 
        "Explain this historical event in an engaging way for teenagers: {question}",
    
    ('history', 'concept_explanation', 'high_school'): 
        "Provide a detailed historical analysis with context and significance: {question}",
    
    # Creative Stories - Language variations
    ('language', 'creative_story', 'elementary'): 
        "Write a short, imaginative story suitable for children: {question}",
    
    ('language', 'creative_story', 'middle_school'): 
        "Create an engaging story that would interest teenagers: {question}",
    
    # Language Help
    ('language', 'language_help', 'elementary'): 
        "Explain this language concept in very simple terms: {question}",
    
    ('language', 'language_help', 'middle_school'): 
        "Provide a clear explanation of this language concept: {question}",
    
    ('language', 'language_help', 'high_school'): 
        "Give a detailed explanation of this language concept: {question}",
}

# Special language-specific templates
hindi_story_template = "हिंदी में एक रोचक कहानी लिखिए: {question}"
hinglish_story_template = "Write a fun story mixing Hindi and English: {question}"

print("✓ Created prompt template system")
print(f"Number of prompt templates: {len(prompt_templates)}")

print("\nSample prompt templates:")
sample_keys = list(prompt_templates.keys())[:3]
for key in sample_keys:
    print(f"  {key}: {prompt_templates[key][:80]}...")
```

```python
# Complete Query Processing System
print("BUILDING COMPLETE QUERY PROCESSING SYSTEM")
print("=" * 50)

def process_student_query(question, vectorizer, classifiers, prompt_templates):
    """
    Process a student query end-to-end:
    1. Classify the query
    2. Select the appropriate prompt template
    3. Return the optimized prompt
    """
    print(f"Processing query: '{question}'")
    print("-" * 40)
    
    # Step 1: Convert question to numbers
    question_vectorized = vectorizer.transform([question])
    
    # Step 2: Get predictions from all classifiers
    age_pred = classifiers['age_group'].predict(question_vectorized)[0]
    subject_pred = classifiers['subject'].predict(question_vectorized)[0]
    language_pred = classifiers['language'].predict(question_vectorized)[0]
    intent_pred = classifiers['intent'].predict(question_vectorized)[0]
    
    print(f"Classification Results:")
    print(f"  Age Group: {age_pred}")
    print(f"  Subject: {subject_pred}")
    print(f"  Language: {language_pred}")
    print(f"  Intent: {intent_pred}")
    
    # Step 3: Select the appropriate prompt template
    prompt = None
    
    # First, check for language-specific story templates
    if intent_pred == 'creative_story':
        if language_pred == 'hindi':
            prompt = hindi_story_template.format(question=question)
        elif language_pred == 'hinglish':
            prompt = hinglish_story_template.format(question=question)
    
    # If no language-specific template found, use the main templates
    if prompt is None:
        template_key = (subject_pred, intent_pred, age_pred)
        if template_key in prompt_templates:
            prompt = prompt_templates[template_key].format(question=question)
        else:
            # Fallback template if no specific template found
            prompt = f"Provide a helpful response to: {question}"
    
    print(f"Selected Prompt: {prompt}")
    print("=" * 50)
    
    return {
        'original_question': question,
        'age_group': age_pred,
        'subject': subject_pred,
        'language': language_pred,
        'intent': intent_pred,
        'optimized_prompt': prompt
    }

print("✓ Query processing function created")

# Test our complete system
print("\nTESTING COMPLETE SYSTEM:")
print("=" * 30)

test_questions = [
    "Why do we have day and night?",
    "How do I solve 3x + 7 = 22?",
    "Write a short story in Hindi about friendship"
]

for i, question in enumerate(test_questions, 1):
    print(f"\nTest {i}:")
    result = process_student_query(question, vectorizer, classifiers, prompt_templates)
    print()
```

#  Feedback Collection System

```python
print("BUILDING FEEDBACK LOOP SYSTEM")
print("=" * 50)

# We'll create a system to track which prompt templates work well and which don't
class FeedbackSystem:
    def __init__(self, prompt_templates):
        self.prompt_templates = prompt_templates
        self.feedback_data = []
        self.template_performance = {}
        
        # Initialize performance tracking for each template
        for template_key in prompt_templates.keys():
            self.template_performance[template_key] = {
                'total_uses': 0,
                'positive_feedback': 0,
                'negative_feedback': 0,
                'success_rate': 0.0
            }
    
    def record_feedback(self, question_data, user_rating, actual_labels=None):
        """
        Record user feedback for a response
        user_rating: 'positive' or 'negative'
        actual_labels: If we know the correct labels (for training data)
        """
        # Extract the template key used
        template_key = (question_data['subject'], question_data['intent'], question_data['age_group'])
        
        # Record the feedback
        feedback_entry = {
            'question': question_data['original_question'],
            'predicted_labels': {
                'age_group': question_data['age_group'],
                'subject': question_data['subject'], 
                'language': question_data['language'],
                'intent': question_data['intent']
            },
            'actual_labels': actual_labels,
            'template_used': template_key,
            'user_rating': user_rating,
            'timestamp': pd.Timestamp.now()
        }
        
        self.feedback_data.append(feedback_entry)
        
        # Update template performance
        if template_key in self.template_performance:
            self.template_performance[template_key]['total_uses'] += 1
            if user_rating == 'positive':
                self.template_performance[template_key]['positive_feedback'] += 1
            else:
                self.template_performance[template_key]['negative_feedback'] += 1
            
            # Calculate success rate
            total = self.template_performance[template_key]['total_uses']
            positive = self.template_performance[template_key]['positive_feedback']
            self.template_performance[template_key]['success_rate'] = positive / total if total > 0 else 0.0
        
        print(f"✓ Feedback recorded: {user_rating} rating for template {template_key}")
    
    def get_template_performance(self):
        """Get performance statistics for all templates"""
        if not self.template_performance:
            return pd.DataFrame()
        
        # Convert the dictionary to DataFrame properly
        performance_list = []
        for template_key, stats in self.template_performance.items():
            performance_list.append({
                'template': template_key,
                'total_uses': stats['total_uses'],
                'positive_feedback': stats['positive_feedback'],
                'negative_feedback': stats['negative_feedback'],
                'success_rate': stats['success_rate']
            })
        
        performance_df = pd.DataFrame(performance_list)
        performance_df = performance_df.sort_values('success_rate', ascending=False)
        return performance_df
    
    def suggest_template_improvements(self):
        """Identify templates that need improvement"""
        performance_df = self.get_template_performance()
        
        print("TEMPLATE PERFORMANCE ANALYSIS:")
        print("=" * 40)
        
        if performance_df.empty:
            print("No performance data available yet")
            return performance_df
        
        # Show best performing templates
        print("\n🏆 BEST PERFORMING TEMPLATES (Success Rate > 80%):")
        best_templates = performance_df[performance_df['success_rate'] > 0.8]
        if len(best_templates) > 0:
            for _, row in best_templates.head(3).iterrows():
                # FIXED: Use the correct column name
                print(f"  {row['template']}: {row['success_rate']:.1%} ({row['total_uses']} uses)")
        else:
            print("  No templates reached 80% success rate yet")
        
        # Show worst performing templates
        print("\n⚠️ TEMPLATES NEEDING IMPROVEMENT (Success Rate < 50%):")
        worst_templates = performance_df[
            (performance_df['success_rate'] < 0.5) & 
            (performance_df['total_uses'] > 2)  # Only show templates with enough data
        ]
        if len(worst_templates) > 0:
            for _, row in worst_templates.iterrows():
                # FIXED: Use the correct column name
                print(f"  {row['template']}: {row['success_rate']:.1%} ({row['total_uses']} uses)")
        else:
            print("  No poorly performing templates identified yet")
        
        return performance_df

# Initialize our feedback system
feedback_system = FeedbackSystem(prompt_templates)

print("✓ Feedback system initialized")
print(f"Tracking {len(feedback_system.template_performance)} prompt templates")
```

```python
# Complete Self-Improving AI Buddy System
print("COMPLETE SELF-IMPROVING AI BUDDY SYSTEM")
print("=" * 50)

class AIBuddySystem:
    def __init__(self, vectorizer, classifiers, prompt_templates):
        self.vectorizer = vectorizer
        self.classifiers = classifiers
        self.prompt_templates = prompt_templates
        self.feedback_system = FeedbackSystem(prompt_templates)
        self.usage_stats = {
            'total_questions_processed': 0,
            'questions_by_subject': {},
            'questions_by_intent': {}
        }
    
    def process_question(self, question):
        """Process a student question end-to-end"""
        self.usage_stats['total_questions_processed'] += 1
        
        print(f" Processing: '{question}'")
        print("-" * 40)
        
        # Get classification
        question_vectorized = self.vectorizer.transform([question])
        
        age_pred = self.classifiers['age_group'].predict(question_vectorized)[0]
        subject_pred = self.classifiers['subject'].predict(question_vectorized)[0]
        language_pred = self.classifiers['language'].predict(question_vectorized)[0]
        intent_pred = self.classifiers['intent'].predict(question_vectorized)[0]
        
        # Update usage statistics
        self.usage_stats['questions_by_subject'][subject_pred] = \
            self.usage_stats['questions_by_subject'].get(subject_pred, 0) + 1
        self.usage_stats['questions_by_intent'][intent_pred] = \
            self.usage_stats['questions_by_intent'].get(intent_pred, 0) + 1
        
        # Select prompt template
        prompt = self._select_prompt_template(subject_pred, intent_pred, age_pred, language_pred, question)
        
        result = {
            'original_question': question,
            'age_group': age_pred,
            'subject': subject_pred,
            'language': language_pred,
            'intent': intent_pred,
            'optimized_prompt': prompt,
            'usage_stats': self.usage_stats.copy()
        }
        
        print(f" Classification: {age_pred} student | {subject_pred} | {intent_pred} | {language_pred}")
        print(f" Optimized Prompt: {prompt}")
        print("=" * 60)
        
        return result
    
    def _select_prompt_template(self, subject, intent, age, language, question):
        """Select the best prompt template based on classification and performance"""
        template_key = (subject, intent, age)
        
        # Check if we have performance data for this template
        if template_key in self.feedback_system.template_performance:
            performance = self.feedback_system.template_performance[template_key]
            if performance['total_uses'] > 0 and performance['success_rate'] < 0.3:
                print(f"  Template {template_key} has low success rate ({performance['success_rate']:.1%})")
        
        # Special handling for language-specific stories
        if intent == 'creative_story':
            if language == 'hindi':
                return hindi_story_template.format(question=question)
            elif language == 'hinglish':
                return hinglish_story_template.format(question=question)
        
        # Use the appropriate template
        if template_key in self.prompt_templates:
            return self.prompt_templates[template_key].format(question=question)
        else:
            return f"Provide a helpful educational response to: {question}"
    
    def record_feedback(self, question_data, user_rating, actual_labels=None):
        """Record user feedback for continuous learning"""
        self.feedback_system.record_feedback(question_data, user_rating, actual_labels)
    
    def get_system_report(self):
        """Generate a comprehensive system report"""
        print(" AI BUDDY SYSTEM REPORT")
        print("=" * 50)
        print(f"Total Questions Processed: {self.usage_stats['total_questions_processed']}")
        print(f"Feedback Entries Collected: {len(self.feedback_system.feedback_data)}")
        
        print("\n USAGE STATISTICS:")
        print("By Subject:")
        for subject, count in self.usage_stats['questions_by_subject'].items():
            print(f"  {subject}: {count}")
        
        print("\nBy Intent:")
        for intent, count in self.usage_stats['questions_by_intent'].items():
            print(f"  {intent}: {count}")
        
        # Performance analysis
        performance_df = self.feedback_system.get_template_performance()
        if not performance_df.empty:
            avg_success_rate = performance_df['success_rate'].mean()
            print(f"\n AVERAGE SUCCESS RATE: {avg_success_rate:.1%}")
        
        return self.usage_stats

# Initialize the complete AI Buddy System
ai_buddy = AIBuddySystem(vectorizer, classifiers, prompt_templates)

print("✓ AI Buddy System Initialized!")
print(" Your educational AI buddy is ready to learn and improve!")

# Test the complete system
print("\n" + "=" * 50)
print("TESTING COMPLETE AI BUDDY SYSTEM")
print("=" * 50)

test_question = "Why do stars twinkle?"
print(f"Student Question: '{test_question}'")

result = ai_buddy.process_question(test_question)

# Simulate positive feedback
ai_buddy.record_feedback(result, 'positive')

print("\n" + "=" * 50)
# Generate system report
ai_buddy.get_system_report()
```
